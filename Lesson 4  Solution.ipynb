In this lesson we will plot data from the Joshua Project about countries in Africa and in Asia with the highest need for Christian workers. The data for this lab is found in the Excel files Africa.xlsx and Asia.xlsx.

```python
import pandas as pd
```

<h3>1. Data for Africa</h3>

1a) Use a command of the form pd.read_excel() to read in the data from Africa.xlsx into a dataframe called Africa_data

```python
Africa_data=pd.read_excel("Africa.xlsx")
```

1b) use the .head() command to display the first 5 rows of Africa_data

```python
Africa_data.head(5)
```

1c) Use the .columns command to list the columns in Africa_data

```python
Africa_data.columns
```

1d)  Create a dataframe called Africa_df which contains the columns ['Population','Abbv','Ctry','UnreachedPop','WorkersNeeded'] from Africa_data. Then display the first 5 rows in Africa_df

```python
Africa_df=Africa_data[['Population','Abbv','Ctry','UnreachedPop','WorkersNeeded']]
Africa_df.head()
```

1e) Note that there is some mussing data for Botswana (NaN). We will replace the missing data with 0's'


```python
Africa_df=Africa_df.fillna (0)
Africa_df.head()
```

1f) Use the .shape command to find the number of rows and columns in Africa_df

```python
Africa_df.shape
```

1g) Save the data to an an Excel file 'AfricaWorkersNeeded' using a command of the form .to_excel()

```python
Africa_df.to_excel('AfricaWorkersNeeded.xlsx')
```

1h) we will import a plot librar and a function to help round off decimals.

```python
import matplotlib.pyplot as plt
from math import trunc
```

1i) We will make a graph which shows the countries in Africa which need more than 500 workers.

```python
plt.figure(figsize=(50,50))
plt.scatter(Africa_df['Population'], Africa_df['UnreachedPop'], c ="red", linewidths = 1, marker ="*", edgecolor ="black", s = Africa_df["WorkersNeeded"])
for i in Africa_df.index:
    if Africa_df.loc[i,'WorkersNeeded']>500:
        plt.text(Africa_df.loc[i,'Population'],Africa_df.loc[i,'UnreachedPop'],Africa_df.loc[i,'Abbv'],size=50)
        plt.text(Africa_df.loc[i,'Population'],Africa_df.loc[i,'UnreachedPop']-2000000,Africa_df.loc[i,'WorkersNeeded'],size=30,color='r',ha='left')
plt.title("African Countries with More than 500 Workers Needed",size=50)
plt.xlabel("Country Population (100 millions)",size=50)
plt.xticks(fontsize=40)
plt.ylabel("Unreached Population (10 millions)",size=50)
plt.yticks(fontsize=40)
plt.savefig("AfrWorkersNeeded.png")
plt.show()
```
